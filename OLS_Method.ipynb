OLS METHOD OF LINEAR REGRESSION

```
#importing Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
#Loading Dataset
dataset=pd.read_csv("pizzaFranchise.csv")
```

```
#Storing each columns values in the variable
X=dataset.iloc[:,0]
Y=dataset.iloc[:,1]
```

```
print(X.head())
print(Y.head())
```

```
plt.style.use('seaborn')
plt.scatter(X,Y,color='orange')
plt.title("Pizza Franchise")
plt.xlabel("annual franchise fee")
plt.ylabel("start up cost ")
plt.show()
```

```
#Finding mean of the values of dataset
X_mean=np.mean(X)
y_mean=np.mean(Y)
```

```
n=0 #numerator
d=0#denomenator
for i in range(len(X)):
  n+=(X[i]-X_mean)*(Y[i]-y_mean)
  d+=(X[i]-X_mean)**2
slope=n/d
c=y_mean-(slope*X_mean)
```

```
print(slope,c)
```

```
y_pred=slope*X +c
plt.scatter(X,Y)
plt.plot(X,y_pred,color="red")
plt.show()
```

```
#plot of regression line
plt.scatter(X,Y)
plt.plot([min(X), max(X)], [min(y_pred), max(y_pred)], color='red')
plt.show()
```

R squared Method for OLS

```
def r2_score(Y,y_pred):
    num = np.sum((Y-y_pred)**2)
    denom = np.sum((Y- Y.mean())**2)
    score = (1- num/denom)
    return score*100
```

```
r2_score(Y,y_pred)
```
